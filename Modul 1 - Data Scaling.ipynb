## Penskalaan Data

### Import dan Read Data
Menggunakan MinMaxScaler, Normalizer, StandardScaler dan RobustScaler

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv("data2.csv")
data = pd.DataFrame(df)

print(data)
```

```python
dF = data.drop(columns="Date")
print(dF)
```

### Visualisasi Data

```python
sns.histplot(df['Duration'], kde=True, color='blue')
plt.title("Duration")
plt.ylabel("Nilai")
plt.show()
```

```python
sns.histplot(df['Pulse'], kde=True, color='blue')
plt.title("Pulse")
plt.ylabel("Nilai")
plt.show()
```

```python
sns.histplot(df['Maxpulse'], kde=True, color='blue')
plt.title("Maxpulse")
plt.ylabel("Nilai")
plt.show()
```

```python
sns.histplot(df['Calories'], kde=True, color='blue')
plt.title("Calories")
plt.ylabel("Nilai")
plt.show()
```

### Scaling

```python
scalerS = StandardScaler()
scalerR = RobustScaler()

dF['Duration_S'] = scalerS.fit_transform(dF[['Duration']])
dF['Duration_R'] = scalerR.fit_transform(dF[['Duration']])

dF['Pulse'] = scalerS.fit_transform(dF[['Pulse']])
dF['Maxpulse'] = scalerS.fit_transform(dF[['Maxpulse']])
dF['Calories'] = scalerS.fit_transform(dF[['Calories']])
```

```python
sns.histplot(dF['Duration_S'], kde=True, color='blue')
plt.title("Duration")
plt.ylabel("Nilai")
plt.show()
```

```python
sns.histplot(dF['Duration_R'], kde=True, color='blue')
plt.title("Duration")
plt.ylabel("Nilai")
plt.show()
```
